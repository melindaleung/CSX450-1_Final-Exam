## Abalone Data Description

Here is the data description of summary statistics of the Abalone Data Set.

```r
# 00. Libraries
library(ggplot2)
```

```r
# 01. Open the data set and relabel the columns
abaloneURL <- "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
abaloneCol <- c('sex','length','diameter','height','whole_weight','shucked_weight','viscera_weight','shell_weight','rings')
abalone <- read.table(abaloneURL,sep=",", col.names = abaloneCol)
head(abalone)
```

```r
# 02. Memory Size & Data Dimensions
object.size(abalone)
dim(abalone)
```

```r
# 03. Data Type
str(abalone)
```

```r
# 04. Summary Statistics
summary(abalone)
```

```r
# 05. Scatterplot
pairs(as.matrix(abalone[,2:9]))
```

```r
# 06.Rings - Variable we want to predict
hist(abalone$rings, freq = FALSE)
lines(density(abalone$rings,2), col = 'red')
```
